```python
import pandas as pd

df = pd.read_excel('poems_merged_output.xlsx')
df.head()
```

```python
df = df.drop("Metadata", axis=1)
df.head()
```

```python
from collections import Counter
import matplotlib.pyplot as plt
import ast
import pandas as pd

def analyze_column(df, column_name):
    all_items = []
    
    for item in df[column_name]:
        # Parse string representation of list
        if isinstance(item, str):
            try:
                parsed = ast.literal_eval(item)
                if isinstance(parsed, list):
                    all_items.extend(parsed)
            except:
                pass
        elif isinstance(item, list):
            all_items.extend(item)
    
    counter = Counter(all_items)
    
    # DataFrame for analysis
    stats_df = pd.DataFrame(counter.items(), columns=['item', 'frequency'])
    stats_df = stats_df.sort_values('frequency', ascending=False).reset_index(drop=True)
    
    # Stats
    print(f"\n{column_name.upper()} Statistics:")
    print(f"Unique items: {len(stats_df)}")
    print(f"Total occurrences: {stats_df['frequency'].sum()}")
    print(f"Mean frequency: {stats_df['frequency'].mean():.2f}")
    print(f"Median frequency: {stats_df['frequency'].median():.2f}")
    
    # Display top items
    display(stats_df.head(20))
    

    fig, axes = plt.subplots(2, 1, figsize=(15, 12))

    # --- Top 15 bar chart ---
    top_15 = stats_df.head(15)
    axes[0].barh(top_15['item'], top_15['frequency'])
    axes[0].set_xlabel('Frequency')
    axes[0].set_title(f'Top 15 {column_name}')
    axes[0].invert_yaxis()
    
    # Add value labels to bars
    for i, (item, freq) in enumerate(zip(top_15['item'], top_15['frequency'])):
        axes[0].text(freq + 0.1, i, f'{freq}', va='center', fontsize=10)

    # --- Frequency Distribution Histogram ---
    axes[1].hist(stats_df['frequency'], bins=30, edgecolor='black')
    axes[1].set_xlabel('Frequency')
    axes[1].set_ylabel('Count of Items')
    axes[1].set_title(f'{column_name} Frequency Distribution')
    
    plt.tight_layout()
    plt.show()


# Analyze
people_stats = analyze_column(df, 'people')
places_stats = analyze_column(df, 'places')
objects_stats = analyze_column(df, 'objects')
```

```python
# Choose columns where uniform case makes sense
columns_to_normalize = ['sentiments', 'people', 'places', 'objects']

for col in columns_to_normalize:
    df[col] = df[col].str.lower()  # or .str.strip().str.lower()
    
df.head()
```
